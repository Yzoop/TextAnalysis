# Text data analysis research


## libraries

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
import spacy
from spacy import displacy
import nltk
```

## Text examples

```python
texts = ["Ambani good to go at Gujrat to start a agro based industry in jio Mart for $70 million",
         "Microsoft is trying to buy France based startup at $7 Million",
         "The quick brown fox, snatch the piece of cube from mouth of black crow",]
```

```python
ui_example = "London has been a key driver in the UK’s FinTech success story. As the second highest ranking FinTech ecosystem globally1, it has the world’s highest concentration of financial and professional services firms. It’s also attractive to investors: in 2020 $4.1 bn, 94% of the UK’s FinTech venture capital, came to the city2."
```

```python
wc = WordCloud(background_color="white",
               max_words=350,
               width=1000,
               height=600,
               random_state=1).generate(ui_example)

plt.figure(figsize=(15,15))
plt.imshow(wc)
plt.axis("off")
```

```python
nlp = spacy.load("en_core_web_md")
```

```python
doc = nlp(ui_example)
```

```python
displacy.render(doc, style="ent", jupyter=True)
```

```python
for entity in doc.ents:
  print(entity)
  print(entity.label_)
  print(str(spacy.explain(entity.label_)))
  print("n")
```

```python
displacy.render(doc, style='dep', jupyter=True, options = {'distance':80})
```
